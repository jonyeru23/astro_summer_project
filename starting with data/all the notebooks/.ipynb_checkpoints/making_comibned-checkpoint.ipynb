```python
import pandas as pd

path = "fig1data_extinction_corrected.txt"


with open(path, "r") as f:
    headers = f.readline().split()
    data = []
    for line in f:
        values = line.split()
        for i, value in enumerate(values):
            if value[-1] == '"':
                values = [values[0]] + [' '.join(values[1:i+1])] + values[i+1:]
                break
        if len(values) != len(headers):
            values = values[:2] + [' '.join(values[2:4])] + values[4:]
        data.append(values)

data = make_numbers(data)

our_data = pd.DataFrame(data, columns=headers)
    
```

```python
def make_numbers(data):
    for row in data:
        for index in [0, 3, 4]:
            row[index] = float(row[index])
    return data
```

```python
def write_to_excel(df, xlsx_path):
    df.to_excel(xlsx_path)
```

# # getting their data

```python
their_data_path = 'their_data.xlsx'
their_data = pd.read_excel(their_data_path, sheet_name="after beaning")

their_data.loc[0, 'V[mag]'] = '19.4 + 0.00'
rel_cols = ['V[mag]', 'FWHM[“]', 'zero point[mag]']
for col in rel_cols:
    for index in range(len(their_data)):
        vals = their_data.loc[index, col].split()
        value, error = float(vals[0]), float(vals[2])
        
        their_data.loc[index, col] = value
        their_data.loc[index, f"error_{col}"] = error
their_data.to_excel(their_data_path, sheet_name="after beaning")
    
```

```python
only_V_o = our_data.loc[our_data['Filter'] == 'V', ['JD_Telescope', 'Magnitude', 'Magnitude_Error']]
t_data = their_data


## setting thetimes right for olny_V_o
only_V_o.loc[:,'JD_Telescope'] -= 2457651.0


only_V_t = t_data.loc[:, ['JD - 2457651.0[day]', 'V[mag]', 'error_V[mag]']]

## setting columns
only_V_o = only_V_o.rename(columns={'JD_Telescope': 'JD - 2457651.0[day]', 'Magnitude': 'V[mag]', 
                                    'Magnitude_Error': 'error_V[mag]'})
all_V = only_V_t.append(only_V_o, ignore_index=True)
all_V
```

```python
import matplotlib.pyplot as plt

x = all_V.loc[:, 'JD - 2457651.0[day]']
y = all_V.loc[:, 'V[mag]']
yerr = all_V.loc[:, 'error_V[mag]']

plt.errorbar(x, y, yerr, fmt='.k', capsize=0)
plt.xlabel('JD - 2457651.0[day]')
plt.ylabel('V[mag]')
plt.gca().invert_yaxis()
plt.show()
```
